```python
# Authenticate your Google account
ee.Authenticate()

# Replace 'your-google-cloud-project-id' with your actual project ID
ee.Initialize(project='landsat-463114')
# Create map
Map = geemap.Map(center=[20.5937, 78.9629], zoom=5)

# Load Landsat 8 Collection 2 Level 2 Surface Reflectance data
image = ee.ImageCollection('LANDSAT/LC08/C02/T1_L2') \
            .filterDate('2021-01-01', '2021-12-31') \
            .filterBounds(ee.Geometry.Point(78.9629, 20.5937)) \
            .sort('CLOUD_COVER') \
            .first()

# Convert surface reflectance to top-of-atmosphere reflectance (scaled)
# L2 bands need scaling by 0.0000275 and offset by -0.2
def scaleL8sr(img):
    opticalBands = img.select(['SR_B2', 'SR_B3', 'SR_B4']).multiply(0.0000275).add(-0.2)
    return opticalBands.copyProperties(img, img.propertyNames())

image_scaled = scaleL8sr(image)

# Visualization parameters
vis_params = {
    'bands': ['SR_B4', 'SR_B3', 'SR_B2'],  # RGB
    'min': 0,
    'max': 0.3,
    'gamma': 1.4
}

Map.centerObject(image, 8)
Map.addLayer(image_scaled, vis_params, 'Landsat 8 True Color (C2 L2)')
Map
```
